```python
import igraph
from igraph import *
```

```python
g = Graph()
g.add_vertices(5)
g.add_edges([(0,1),(0,3),(1,2),(1,4),(2,4)])
```

```python
print("Number of vertices in the graph:", g.vcount())
print("Number of edges in the graph: ", g.ecount())
print("List of edges in the graph: ", g.get_edgelist())
print("Density of the graph:", 2*g.ecount()/(g.vcount()*(g.vcount()-1)))
print("Diameter of the graph:", g.diameter())
print("Is the graph directed:", g.is_directed())
print("Degree of each node in the graph",g.degree())
print("Maximum degree in the graph:", g.maxdegree())
print("Betweenness of each node in the graph: ",g.betweenness())
print("Edge Betweenness of each node in the graph: ",g.edge_betweenness())
print("Adjacency matrix:\n", g.get_adjacency())
print("Shortest path from node 0: ",g.get_shortest_paths(0))
print("Shortest path from node 0 to node 4" ,g.get_shortest_paths(0, to=4))
```

```python
import matplotlib.pyplot as plt
```

```python
ga=Graph()
ga.add_vertices(10)
ga.add_edges([(0,1),(0,2),(0,3),(1,3),(2,3),(2,4),(4,5),(4,8),(4,9),(5,6),(8,9),(5,7),(6,7)])
```

```python
eb=ga.community_edge_betweenness()
clusters =eb.as_clustering()
membership = clusters.membership
```

```python
ga.vs["name"] = ["A","B","C","D","E","F","G","H","I","J"]
ga.vs["gender"] = ["f","m","f","m","f","f","m","m","m","f"]
ga.vs["membership"]=membership
ga.es["friendship"] = [False, False, True, True, True, False, False, True, False, True, True, False, True]
ga.vs["label"] = ga.vs["name"]
```

```python
layout = ga.layout_drl()
plot(ga,layout=layout, vertex_color='grey',bbox=(200, 200), margin=20)
```

```python
layout = ga.layout_circle()
plot(ga,layout=layout, vertex_color='grey',bbox=(200, 200), margin=20)
```

```python
layout = ga.layout_kamada_kawai()
plot(ga,layout=layout, vertex_color='grey',bbox=(200, 200), margin=20)
```

```python
layout = ga.layout_fruchterman_reingold()
plot(ga,layout=layout, vertex_color='grey',bbox=(200, 200), margin=20)
```

```python
layout = ga.layout_kamada_kawai()
color= {"m": "lavender", "f": "rosybrown1"}
plot(ga, layout=layout, vertex_color=[color[gender] for gender in ga.vs["gender"]], bbox=(200, 200), margin=20)
```

```python
layout = ga.layout_kamada_kawai()
plot(ga, layout=layout, edge_width = [2 + 2 * int(friendship) for friendship in ga.es["friendship"]],vertex_color='grey', bbox=(200, 200), margin=20)
```

```python
layout = ga.layout_kamada_kawai()
color= {"0": "darkseagreen1", "1": "wheat1","2":"plum1"}
plot(ga, layout=layout,vertex_color=[color[str(membership)] for membership in ga.vs["membership"]], bbox=(200, 200), margin=20)
```
